# FEATURE ENGINEERING

> Making the features names more interpretable:

```python
data = data.rename(columns={"Dt_Customer":"Date of Joining", "MntWines":"Wines", "MntFruits":"Fruits", "MntMeatProducts":"Meat", "MntFishProducts":"Fish", "MntSweetProducts":"Sweets", "MntGoldProds": "Gold"})
```

The columns 'MntWines' i.e. the amount spent on wines has been renamed to 'Wines', similar engineering has been done on the other 'Mnt' features as well.

> Converting 'Date of Joining' to 'DaysSinceJoined'

```python
data
```

```python
data.info()
```

```python
data["Date of Joining"] = pd.to_datetime(data["Date of Joining"])     #Converting the type to datetime
```

```python
dates = []
for i in data["Date of Joining"]:
  i = i.date()
  dates.append(i)       # Joining Dates of the newest and oldest recorded customer
print("The newest customer's joining date in the records:",max(dates))
print("The oldest customer's joining date in the records:",min(dates))
```

```python
days = []
recent = max(dates)   #Gives the date of the most recent customer
for i in dates:
    delta = recent - i
    days.append(delta)
data["DaysSinceJoined"] = days
data["DaysSinceJoined"] = pd.to_numeric(data["DaysSinceJoined"], errors="coerce")
```

```python
data.head()
```

```python
data = data.drop("Date of Joining",axis = 1)
```

> Deriving 'Age' of the customers as of now from 'Year_birth

```python
#Age of customer today 
data["Age"] = 2022-data["Year_Birth"]
```

```python
data = data.drop("Year_Birth",axis=1)
```

> Deriving the 'Total Spendings' of the customers by summing
'Wines,Fruits,Meat,Fish,Sweets,Gold'

```python
data["Total Spendings"] = data['Wines']+data['Fruits']+data['Meat']+data['Fish']+data['Sweets']+data['Gold']
```

> Deriving 'Living_with' as 'Partner' or 'Alone' from the feature 'Marital_Status'

```python
data["Living_With"]=data["Marital_Status"].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone",})
```

```python
data = data.drop("Marital_Status",axis=1)
```

> Creating a feature 'Children' that indicates the total children living in the household

```python
data["Children"]=data["Kidhome"]+data["Teenhome"]
```

> Creating a feature for total members in the householde

```python
data["Family_Size"] = data["Living_With"].replace({"Alone": 1, "Partner":2})+ data["Children"]
```

> Creating a feature 'Is_Parent' pertaining parenthood

```python
data["Is_Parent"] = np.where(data.Children>0,1,0)
```

> Segmenting education levels in three groups

```python
data["Education"] = data["Education"].replace({"Basic":"Undergraduate","2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
```

> Dropping redundant columns/features

```python
to_drop = ["Z_CostContact", "Z_Revenue", "ID"]
data = data.drop(to_drop, axis=1)
```

> Getting a list of the Categorical Variables

```python
s = (data.dtypes == 'object')     #The datatype of a categorical variable is object
object_cols = list(s[s].index)
print("Categorical variables in the dataset:", object_cols)
```

> Label Encoding the Categorical Variables as found above

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE=LabelEncoder()
for i in object_cols:
    data[i]=data[[i]].apply(LE.fit_transform)
print("All features are now numerical now")
```

```python
#Creating a copy of the data
data_copy = data.copy()
```

```python
# creating a subset of dataframe by dropping the features on deals accepted and promotions
cols_del = ['AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'AcceptedCmp1','AcceptedCmp2', 'Complain', 'Response']
data_copy = data_copy.drop(cols_del, axis=1)
```

> Performing Scaling using StandardScaler()

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(data_copy)
scaled_data = pd.DataFrame(scaler.transform(data_copy),columns= data_copy.columns )
print("All features are now scaled")
```

```python
#Scaled data to be used for reducing the dimensionality
print("Dataframe to be used for further modelling:")
scaled_data.head()
```

> Using Principal Component Analysis to reduce the dimensionality

We will be initiating PCA to reduce the features to 3

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
PCA_data = pd.DataFrame(pca.transform(scaled_data), columns=(["col1","col2", "col3"]))
PCA_data.describe().T
```

```python
#Visualisation of the Dataset with it's dimension reduced
x =PCA_data["col1"]
y =PCA_data["col2"]
z =PCA_data["col3"]
#To plot
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(x,y,z, c="maroon", marker="o" )
ax.set_title("A 3D Projection Of Data In The Reduced Dimension")
plt.show()
```

# MODEL FITTING

> Finding the optimum number of clusters to make using the 'Elbow Method'

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
print('Elbow Method to determine the number of clusters to be formed:')
Elbow_M = KElbowVisualizer(KMeans(), k=10)
Elbow_M.fit(PCA_data)
Elbow_M.show()
```

Hence, we will be selecting k=5

> Implementing the Agglomerative Clustering Model

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=5)
```

```python
#Fitting the model
y_cluster = ac.fit_predict(PCA_data)
PCA_data["Cluster"] = y_cluster
```

> Assigning the clusters to the whole dataset based on the clustering done above

```python
#Adding the clusters to the original dataframe
data ["Cluster"] = y_cluster
```

```python
data.head()
```

```python
#Visualising the clusters that each datapoint was assigned to
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(x, y, z, s=40, c=PCA_data["Clusters"], marker='o')
ax.set_title("The Plot Of The Clusters")
plt.show()
```

Now, we can use the clusters formed to enquire about any feature, group of features as well.

> Some Examples for the analysis that can be done with the help of Clustering

```python
#Plotting the number of deals purchased
plt.figure()
pl=sns.boxenplot(y=data["NumDealsPurchases"],x=data["Clusters"], palette= 'hls')
pl.set_title("Number of Deals Purchased")
plt.show()
```

```python
Personal = [ "DaysSinceJoined", "Age", "Children", "Family_Size", "Is_Parent", "Education","Living_With"]

for i in Personal:
    plt.figure()
    sns.jointplot(x=data[i], y=data["Total Spendings"], hue =data["Clusters"], palette='hls')
    plt.show()
```
